Normality Test: A test to see whether a data set is well-modeled by a normal distribution and to compute how likely it is for a random variable underlying the data set to be normally distributed.

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import shapiro
```

#### Data input.

Small

```python
LengthS=np.array([3.33,3.125,2.94,2.515,3.105,2.84,2.95,3.22,2.65,2.68])
MicroS=np.array([35,26,12,13,21,46,29,12,19,21])
```

```python
fig,ax=plt.subplots(figsize=(7,3))
ax.scatter(LengthS,MicroS)
ax.set_ylim([0,50])
ax.set_xlabel('Length')
ax.set_ylabel('Number of Microplastics')
plt.show()
```

Medium

```python
LengthM=np.array([3.45,3.95,4.05,4.3,4.4,4.4,4.4,4.4,4.25,4.5])
MicroM=np.array([30,40,26,18,29,30,20,43,36,32])
```

```python
fig,ax=plt.subplots(figsize=(7,3))
ax.scatter(LengthM,MicroM)
ax.set_ylim([0,50])
ax.set_xlabel('Length')
ax.set_ylabel('Number of Microplastics')
plt.show()
```

Large

```python
LengthL=np.array([4.7,4.85,4.45,4.7,4.55,4.5,4.8,4.6,4.9,5.5])
MicroL=np.array([33,33,37,34,48,46,29,35,34,34])
```

```python
fig,ax=plt.subplots(figsize=(7,3))
ax.scatter(LengthL,MicroL)
ax.set_ylim([0,50])
ax.set_xlabel('Length')
ax.set_ylabel('Number of Microplastics')
plt.show()
```

The data is subjectively small, hence why I used Shapiro-Wilk's test instead of Kolmogorov's test.

```python
def shapirowilk(data):
    stat,p=shapiro(data)
    confi=0.05
    if p>confi:
        print(f'Sample looks Gaussian, with p-value = {p}')
    else:
        print(f'Sample does not look Gaussian, with p-value = {p}')
```

If $p<0.05$, then the sample looks Gaussian (normally distributed).

For the small dataset,

```python
shapirowilk(MicroS)
```

For the medium dataset,

```python
shapirowilk(MicroM)
```

For the large dataset,

```python
shapirowilk(MicroL)
```

Hence, both the dataset for medium and small lengths are normally distributed, whereas the large lengths are not.
