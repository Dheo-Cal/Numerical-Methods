```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from pymatreader import read_mat
from scipy.optimize import curve_fit
```

```python
data = read_mat('Data.mat')
newdata = pd.DataFrame(data['Data'])
newdata = newdata.to_numpy()
times = newdata[:,0]
activity = newdata[:,1]
```

```python
plt.subplots(figsize=(7,3),dpi=200)
plt.plot(times,activity,'ko')
plt.xlabel("Time (s)")
plt.ylabel("Activity (count/s)")
plt.show()
```

```python
def ModelFunction(x,a,b):
    return a * np.exp(-b * x)
```

```python
popt,pcov = curve_fit(ModelFunction, times, activity)
```

```python
plt.subplots(figsize=(7,3),dpi=200)
plt.plot(times,activity,'ko')
plt.plot(times,ModelFunction(times,popt[0],popt[1]),'r--')
plt.xlabel("Time (s)")
plt.ylabel("Activity (count/s)")
plt.show()
```
